```python
import pandas as pd
import re

df = pd.read_csv(r"..\data\interim\job_offers.csv", sep=',', encoding='UTF-8', decimal='.')
# Change column names to lowercase
df.columns = df.columns.str.lower()
df
```

```python
# Delete items without salary ranges provided
df = df.dropna(subset=['salary']).copy()
df
```

```python
# Data cleaning
df = df[~df['location'].str.contains(r'\d+ lokalizacj')].copy()
```

```python
print(df['salary'].unique())
df['salary'] = df['salary'].str.replace('\xa0', ' ', regex=False)
print(df['salary'].unique())
```

```python
# Removing hard spaces in 'Salary' column
df['salary'] = df['salary'].str.replace('\xa0', ' ', regex=False)

# Function to separate data
def extract_salary_info(salary_str):
    if not salary_str or pd.isna(salary_str):
        return None, None, None, None, None

    match = re.match(r"(\d[\d\s]*)–(\d[\d\s]*)\s([a-zA-Zżł]+)\s(netto|brutto)\s?(?:\(\+\s?VAT\)|\(zal\. od umowy\))?\s/\s(\w+)|(\d[\d\s]*)\s([a-zA-Zżł]+)\s(netto|brutto)\s?(?:\(\+\s?VAT\))?\s/\s(\w+)"
, salary_str)
    
    if match:
        if match.group(1):  
            min_salary = match.group(1).replace(' ', '')
            max_salary = match.group(2).replace(' ', '')
            currency = match.group(3)
            netto_brutto = match.group(4)
            period = match.group(5)
        else:  
            min_salary = match.group(6).replace(' ', '')
            max_salary = None
            currency = match.group(7)
            netto_brutto = match.group(8)
            period = match.group(9)

        return min_salary, max_salary, currency, netto_brutto, period
    print("Not Matched")
    return None, None, None, None, None

# Applying the function to the Salary column
df[['min_salary', 'max_salary', 'currency', 'netto/brutto', 'period']] = df['salary'].apply(
    lambda x: pd.Series(extract_salary_info(x))
)

# Convert min/max salary to numbers
df['min_salary'] = pd.to_numeric(df['min_salary'], errors='coerce')
df['max_salary'] = pd.to_numeric(df['max_salary'], errors='coerce')

# Clearing empty values
df = df.dropna(subset = ['netto/brutto'])
df
```

```python
# Clearing data
print(df['currency'].unique())
df = df.drop(columns=['currency'])
df
# Check which rows have "godz" in 'period' column
mask = df['period'] == 'godz'

#Multiplying the values ​​in the 'min salary' and 'max salary' columns by 168, only where 'period' == 'godz'
df.loc[mask, 'min_salary'] = df.loc[mask, 'min_salary'] * 168
df.loc[mask, 'max_salary'] = df.loc[mask, 'max_salary'] * 168

# Replace the value in the 'period' column with 'month' where 'period' == 'hour'
df.loc[mask, 'period'] = 'mies'
df
```

```python
# Gross to net conversion factor assumed for the exercise purposes
brutto_to_netto_factor = 0.7  # Assuming that net is 70% gross

# Gross to Net Conversion
def convert_to_netto(row):
    if 'netto/brutto' in row and row['netto/brutto'] == 'brutto':
        row['min_salary'] = row['min_salary'] * brutto_to_netto_factor
        row['max_salary'] = row['max_salary'] * brutto_to_netto_factor
        row['netto/brutto'] = 'netto'  # Change status to 'net'
    return row

# Filling in missing max_salary
df['max_salary'] = df['max_salary'].fillna(df['min_salary'])

# Applying a function to a DataFrame
df = df.apply(convert_to_netto, axis=1)

df = df.rename(columns = {'netto/brutto': 'netto'})
df
```

```python
# Adding the value of average wages
df['salary_avg'] = (df['max_salary'] + df['min_salary'])/2
df
```

```python
df['is_senior'] = df['title'].str.contains('Senior', case=False).astype(int)
df
```

```python
# Cleaning and unifying the syntax of lower and upper case letters
def normalize_location(location):
    location = location.split(',')[0]
    location = location.strip().capitalize()
    return location

df['location'] = df['location'].apply(normalize_location)
df
```

```python
# Save to csv
output_csv = r"..\data\processed\job_offers_transformed.csv"
df.to_csv(output_csv, encoding='UTF-8', index=False)
```
